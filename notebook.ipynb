### Cleaning Data

##### Cleaning "order_reviews"

```python
#Menampilkan kolom yang memiliki missing value
order_reviews_df.isna().sum()
```

```python
# Menampilkan persebaran data
order_reviews_df.review_comment_title.value_counts()
```

```python
# Metode imputasi (dengan modus) untuk mengatasi missing values
modus_review_comment_title = order_reviews_df['review_comment_title'].mode()[0]
order_reviews_df['review_comment_title'].fillna(modus_review_comment_title, inplace=True)
```

Karena data bertipe kategorikal dan adanya kategori yang dominan, digunakan teknik imputasi untuk mengisi data yang kosong dengan modus

```python
# Cek kolom review_comment_title apakah masih terdapat missing values
order_reviews_df.isna().sum()
```

```python
# Menampilkan data(baris) yang review_comment_message-nya kosong atau missing value
order_reviews_df[order_reviews_df.review_comment_message.isna()]
```

```python
# Menampilkan persebaran data kolom review_comment_message
order_reviews_df.review_comment_message.value_counts()
```

```python
# Metode imputasi untuk mengatasi missing values
modus = order_reviews_df['review_comment_message'].mode()[0]
order_reviews_df['review_comment_message'].fillna(modus, inplace=True)
```

Karena data bertipe kategorikal dan adanya kategori yang dominan, digunakan teknik imputasi untuk mengisi data yang kosong dengan modus

```python
# Cek kolom review_comment_message apakah masih terdapat missing values
order_reviews_df.isna().sum()
```

```python
# ganti tipe data
datetime_columns = ["review_creation_date", "review_answer_timestamp"]

for column in datetime_columns:
  order_reviews_df[column] = pd.to_datetime(order_reviews_df[column])
```

```python
order_reviews_df.info()
```

##### Cleaning "orders"

```python
# Ganti tipe data
datetime_columns = ["order_approved_at", "order_delivered_carrier_date", "order_purchase_timestamp", "order_delivered_customer_date", "order_estimated_delivery_date"]

for column in datetime_columns:
  orders_df[column] = pd.to_datetime(orders_df[column])
```

```python
# cek tipe data setelah diubah
orders_df.info()
```

```python
orders_df.isna().sum()
```

```python
# Menampilkan data(baris) yang order_approved_at-nya kosong atau missing value
orders_df[orders_df.order_approved_at.isna()]
```

```python
# Metode imputasi (dengan mean) untuk mengatasi missing value pada order_approved_at
mean_approved_at = orders_df['order_approved_at'].mean()

# Isi missing value dengan nilai rata-rata
orders_df.fillna({'order_approved_at': mean_approved_at}, inplace=True)
```

```python
# Menampilkan data(baris) yang order_delivered_carrier_date-nya kosong atau missing value
orders_df[orders_df.order_delivered_carrier_date.isna()]
```

```python
# isi missing value dengan nilai (tanggal) terakhir yang ada
orders_df.fillna({'order_delivered_customer_date': orders_df['order_delivered_customer_date'].max()}, inplace=True)
```

```python
# Menampilkan data(baris) yang order_delivered_carrier_date-nya kosong atau missing value
orders_df[orders_df.order_delivered_customer_date.isna()]
```

```python
orders_df.isna().sum()
```

**Insight:**
Pengubahan tipe data dan penanganan missing value pada tabel orders telah berhasil dilakukan

### Pertanyaan 3: Bagaimana penilaian pelanggan terhadap pembelian?

### Pertanyaan 6: Bagaimana jumlah order pada setiap bulan?

```python
# mengubah format kolom  order_approved_at dari datetime ke yearmonth
monthly_orders_df = all_df.resample(rule='ME', on='order_approved_at').agg({
    "order_id": "nunique",
    "price": "sum"
})
monthly_orders_df.index = monthly_orders_df.index.strftime('%Y-%m')
monthly_orders_df = monthly_orders_df.reset_index()
monthly_orders_df.rename(columns={
    "order_id": "order_count",
    "price": "revenue"
}, inplace=True)
monthly_orders_df
```

```python
monthly_orders_df = all_df.resample(rule='ME', on='order_approved_at').agg({
    "order_id": "nunique",
    "price": "sum"
})
monthly_orders_df.index = monthly_orders_df.index.strftime('%B %Y')

monthly_orders_df = monthly_orders_df.reset_index()
monthly_orders_df.rename(columns={
    "order_id": "order_count",
    "price": "revenue"
}, inplace=True)

plt.figure(figsize=(15, 5))
plt.plot(monthly_orders_df["order_approved_at"], monthly_orders_df["order_count"], marker='o', linewidth=2, color="#72BCD4")
plt.title("Orders per Month (2016-2018)", loc="center", fontsize=20)
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```
